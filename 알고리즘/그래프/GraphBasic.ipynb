### 그래프(Graph)
- 그래프는 실제 세계의 현상이나 사물을 정점(Vertex) 또는 노드(Node) 와 간선(Edge) 로 표현하기 위해 사용한다.

#### 그래프 관련 용어
* 중요 용어
- 노드(Node) : 현재 위치를 말한다, 정점(Vertex) 이라고도 한다.
- 간선(Edge) : 위치 간의 관계를 표시한 선으로 노드를 연결한 선 이라고 보면 된다.(link 또는 branch 라고도 한다.)

* 참고 용어
- 인접 정점(Adjacent Vertex) : 간선으로 직접 연결된 정점(또는 노드)
- 정점의 차수(Degree) : 무방향 그래프에서 하나의 정점에 인접한 정점의 수
- 진입 차수(In-Degree) : 방향 그래프에서 외부에서 오는 간선의 수
- 진출 차수(Out-Degree) : 방향 그래프에서 외부로 향하는 간선의 수
- 경로 길이(Path Length) : 경로를 구성하기 위해 사용된 간선의 수
- 단순 경로(Simple Path) : 처음 정점과 끝 정점을 제외하고 중복된 정점이 없는 경로(처음과 끝이 중복되는 노드일 경우는 단순 경로로 허용한다.)
- 사이클(Cycle) : 단순 경로의 시작 정점과 종료 정점이 동일한 경우

#### 그래프 종류
- 무방향 그래프
    - 방향이 없는 그래프
    - 노드는 간선을 통해서 양방향으로 갈 수 있다.
    - 보통 노드 A,B 가 연결되어 있을 경우 (A,B) 또는 (B,A) 로 표기한다.

- 방향 그래프
    - 간선에 방향이 있는 그래프
    - 보통 노드 A, B 가 A -> B 로 가는 간선으로 연결되어 있을 경우, <A,B> 로 표기
    - (<B,A> 는 B -> A 로 가는 간선이 있는 경우이므로 <A,B> 와는 다르다.)
    
- 가중치 그래프 or 네트워크
    - 간선에 비용 또는 가중치가 할당된 그래프

- 연결 그래프와 비연결 그래프
    - 연결 그래프
        - 무방향 그래프에 있는 모든 노드에 항상 경로가 존재하는 경우
    - 비연결 그래프
        - 무방향 그래프에서 특정 노드에 대해 경로가 존재하지 않는 경우
        
- 사이클 과 비순환 그래프
    - 사이클
        - 단순 경로의 시작 노드와 종료 노드가 동일한 경우
    - 비순환 그래프
        - 사이클이 없는 그래프
        
- 완전 그래프
    - 그래프의 모든 노드가 서로 연결되어 있는 그래프


#### 그래프와 트리의 차이
- 트리는 그래프 중에 속한 특별한 종류라고 볼 수 있다.
- 정의:
    - 그래프 :노드와 노드를 연결하는 간선으로 표현되는 자료 구조
    - 트리 : 그래프의 한 종류, 방향성이 있는 비순환 그래프
- 방향성:
    - 그래프 : 방향 그래프, 무방향 그래프 둘 다 존재한다.
    - 트리 : 방향 그래프만 존재한다.
- 사이클:
    - 그래프 : 사이클이 구성되는 것이 가능하다. 순환 및 비순환 그래프 모두 존재한다.
    - 트리 : 비순환 그래프로 사이클이 존재하지 않는다.
- 루트 노드:
    - 그래프 : 루트 노드가 존재하지 않는다.
    - 트리 : 루트 노드가 존재한다.
- 부모/자식 관계:
    - 그래프 : 부모 자식 개념이 존재하지 않는다.
    - 트리 : 부모 자식 관계가 존재한다.
